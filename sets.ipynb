Set :
it is a unordered collection of items.
Sets can also be used to perform mathematic set operations like union , intersection,symmetric difference etc.

Characteristics :
Unordered
Mutable
No Duplicate
Can't contain mutable data types

```python
# Creating Sets

# empty set
s =set()
print(s)
print(type(s))
```

```python
# ID set
s = {1,2,3}
print(s)
print(type(s))
```

```python
# 2D sets
s = {1,2,3,{4,5}}
print(s)
```

```python
# homo and hetro
s = {1,'hello',2.3,True,(7,8,9)}
print(s)
```

```python
# using type coversion :
s = set([1,2,3,4])
print(s)
```

```python
# duplicate not allowed
s = {1,1,2,2,3,3}
print(s)
```

```python
# set can't have mutable items
s = {[1,2,3,4]}
print(s)
```

```python
# order never matterns in Stes
s1 = {1,2,3}
s2 = {3,2,1}
print(s1 == s2)
```

```python
# Accessing items :

s ={1,2,3,4}
s1[0:2] # not allowed accessing items
```

```python
#Editing Items :
s = {1,2,3,4}
s[0] = 500 # editing is not allowed becoz sets are inmutable 
```

```python
# Adding Items :
s = {1,2,3,4}
# add
s.add(5)
print(s)
# update
s.update([5,6,7,8])
print(s)
```

```python
# Deleting items
s = {1,2,3,4}
print(s)
del s
#print(s) # deleted
```

```python
# Discard :
s = {1,2,3,4,5}
print(s)
s.discard(5)
print(s)

# remove :
s.remove(3)
print(s)

#pop
print(s) #in sets pop delete items randomly
s.pop()
```

Sets Operations :
1. union(|)
2. intersection(&)
3. Difference(-)
4. Symmetric Difference(^)
5. Membership test
6. Iteration

```python
# Union
s1 = {1,2,3,4,4}
s2 = {4,5,6,7,8}
s1 | s2
```

```python
# Intersection :
s1 & s2
```

```python
# Difference (-)
s1 - s2
```

```python
# Symmetric difference(^)
s1 ^ s2
```

```python
# Membership
1 in s1 
1 not in s1
```

```python
# Iteration
for i in s1 :
    print(i)
```

```python
# Len/sum/min/max/sorted :
#len()
s = {3,1,2,5,6,8,9}
print(len(s))

```

```python
# sum :
print(sum(s))
```

```python
# min :
print(min(s))
```

```python
# max :
print(max(s))
```

```python
# sorted :
print(sorted(s))
print(s)
      
```

```python
# union / upadate :
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
# s1 / s2
s1.union(s2)
s1.update(s2)
print(s1)
print(s2)
```

```python
# Intersection / Intersection_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
s1.intersection(s2)
s1.intersection_update(s2)
print(s1)
print(s2)
```

```python
# Diffrence / Difference-update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
s1.difference(s2)
s1.difference_update(s2)
print(s1)
print(s2)
```

```python
# Symmetric_difference / symmetric_didderence_update
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
s1.symmetric_difference(s2)
s1.symmetric_difference_update(s2)
print(s1)
print(s2)
```

```python
# isdisjoint/issubset/issuperset
s1 = {1,2,3,4,5}
s2 = {4,5,6,7,8}
s1.isdisjoint(s2)
```

```python
s1 = {1,2,3,4,5}
s2 = {4,5}
s1.issubset(s2)
```

```python
s1 = {1,2,3,4,5}
s2 = {4,5}
s1.issuperset(s2)
```

```python
s1 = {1,2,3,4,5}
s2 = s1.copy()
print(s1)
print(id(s1))
print(s2)
print(id(s2))
```
