Zip :
the zip() function returns a zip object,which is an iterator of tuples where the first item in each passed iterator in paired together,and then the second item in each passed iterator are paired together.

```python
l1 = [1,2,3,4]
l2 = [-1,-2,-3,-4]
list(zip(l1,l2))
[i+j for i,j in zip(l1,l2)]
```

Difference b/w Lists and Tuple
1.Syntax
2.Mutability
3.Speed
4.Merory
5.Built in Functionlity
6.Error prone
7.Usability

```python
import time
l = list(range(10000000))
t = tuple(range(10000000))
start = time.time()
for i in l:
    i * 5
print('list time',time.time()-start)
start = time.time()
for i in t:
    i * 5
print('tuple time',time.time()-start)
```

```python
import sys
l = list(range(1000))
t = tuple(range(1000))
print('list size',sys.getsizeof(l))
print('tuple size',sys.getsizeof(t)) # tuple is faster then list
```
