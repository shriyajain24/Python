```python
class Restaurant:
    def __init__(self):
        print("--- Welcome to our Restaurant ---")
        self.name = input(" Aapka Naam: ")
        self.menu = {"Chai": 5, "Samosa": 10, "Coffee": 30,"Paneer Paratha": 80}
        self.bill = 0
        self.selection() 

    def selection(self):
        print("\n1. Menu | 2. Order | 3. Bill | 4. Exit")
        choice = input(" Hello mai Aapki ky madad kar sakta hu ( enter choice 1 to 4) : ")

        if choice == '1':
            self.show_menu()
        elif choice == '2':
            self.take_order()
        elif choice == '3':
            self.show_bill()
        elif choice == '4':
            print(" Shukriya! Phir aaiyega.")
            return
        else:
            print(" Galat number! Dobara chunein.")
            self.selection()

    def show_menu(self):
        print(" \n--- MENU --- ")
        for dish in self.menu:
            print(f"{dish}: Rs.{self.menu[dish]}")
        self.selection() 

    def take_order(self):
        item = input( " \n Aap ky order karna chate hai : ").title()
        if item in self.menu:
            qty = int(input(" Quantity : "))
            self.bill = self.bill + (self.menu[item] * qty)
            print(f"{item} add ho gaya!")
        else:
            print(" Sorry, ye dish nahi hai.")
        self.selection()

    def show_bill(self):
        print(f"\nTotal Bill: Rs.{self.bill}")
        self.selection()


hotel = Restaurant()
```
