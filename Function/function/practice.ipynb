```python
# 1. Encapsulation: Menu items ki details safe rakhi hain
class MenuItem:
    def __init__(self, name, price):
        self.name = name
        self.__price = price  # Private variable

    def get_price(self):
        return self.__price

# 2. Aggregation: Restaurant 'has-a' Menu
class Restaurant:
    def __init__(self, restaurant_name):
        self.restaurant_name = restaurant_name
        self.menu = []  # List of MenuItem objects

    def add_item_to_menu(self, item):
        self.menu.append(item)
        print(f"{item.name} ko menu mein add kar diya gaya hai.")

    def show_menu(self):
        print(f"\n--- {self.restaurant_name} Ka Menu ---")
        for item in self.menu:
            print(f"{item.name}: ₹{item.get_price()}")

    # 3. Polymorphism: Is method ko child class override karegi
    def welcome_message(self):
        print(f"Welcome to {self.restaurant_name}!")

# 4. Inheritance: Restaurant se FastFoodRestaurant banaya
class FastFoodRestaurant(Restaurant):
    def __init__(self, name, delivery_time):
        super().__init__(name) # Parent ka constructor call kiya
        self.delivery_time = delivery_time

    # Method Overriding (Polymorphism)
    def welcome_message(self):
        print(f"Bhaiya jaldi aao! Welcome to {self.restaurant_name}. Delivery in {self.delivery_time} mins!")

# --- Execution (Commands) ---

# Restaurant ka object banaya
my_cafe = FastFoodRestaurant("Zayka Express", 20)

# Menu Items banaye
burger = MenuItem("Veg Burger", 120)
pizza = MenuItem("Cheese Pizza", 250)
chai = MenuItem("Masala Chai", 30)

# Commands chala rahe hain
my_cafe.welcome_message() # Overridden method chalega
my_cafe.add_item_to_menu(burger)
my_cafe.add_item_to_menu(pizza)
my_cafe.add_item_to_menu(chai)

my_cafe.show_menu()
```
