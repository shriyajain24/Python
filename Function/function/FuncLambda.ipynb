```python
# args and kwargs
# they are special python keyword that are used to pass the variable lenght of arg in a functn
# allow us to pass a variable no of non - keyword

def multi(*args): # variable args (it make internal tuple)
    a = 1
    for i in args:
        a = a * i
    print(args)
    return a
```

```python
multi(2,3,5,6,7)
```

```python
# **kwargs
# **kwargs allows us to pass number of keywords argument
# keyword arguments mean that they contain a key - value pair , like a python dictionary.

def display(**kwargs):
    for(key,value) in kwargs.items():
        print(key,'->',value)
```

```python
display(india='dehli',nepal='katmandu',pakistan='islamabad')
```

# Lambda Function :
   A lambda function is a small ananymous function
   A lambda function can take any number of arguments,but can only have one expression

Example :
    lambda a,b : a + b

Syntax :
    lambda parameter : expression

```python
# x,y -> x ^ y
a = lambda x:x**2
print(a(2))
```

```python
# x,y -> x + y
a = lambda x,y : x*y
print(a(2,3))
```

Diff b/w lambda v/v Normal Function :
1.No name
2.lambda has no return value(infact,return a function)
3.lambda is written in 1 line
4.not reuable

then why use lambda function
they are used in HOF(higher order function)

```python
# check if a string has 'a'
a = lambda s : 's' in s
a('shri')
```

```python
# odd or even
a = lambda x : 'even' if x % 2 == 0 else 'odd'
print(a(6))
```

Higher order Function :

```python
# example

def sq(x):
    return x**2
def cube(x):
    return x**3

def transform(f,l):
    output = []
    for i in l:
        output.append(f(i))
    print(output)
l =[1,2,3,4,5]
transform(lambda x:x**3,l)
```

```python
# fetch names from a list of dict :
users = [
    {
        'name':'shri',
        'age':20,
        'gender':'female'
    },
    {
        'name':'shriya',
        'age':20,
        'gender':'female'
    },
    {
        'name':'diya',
        'age':19,
        'gender':'female'
    }

]
res = list(map(lambda user:user['gender'],users))
print(res)
```

```python
# Filter

# number greater than s
l =[3,4,2,7,8,9]
list(filter(lambda x:x>5,l))
```
