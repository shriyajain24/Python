```python
class Point:

    def __init__(self , x , y):
        self.x = x
        self.y = y
    
    def __str__(self):
        return "<{}/{}>".format(self.x,self.y)
    
    def euclidean_distance(self,other):
        return( (self.x - other.x)**2 + (self.y - other.y)**2)**0.5
    
    def distance_from_origin(self):
        #return(self.x**2 + self.y**2)**0.5
        return self.euclidean_distance(Point(0,0))
    



```

```python
class line:
    def __init__(self,a , b , c):
        self.a = a
        self.b = b
        self.c = c
    def __str__(self):
        return '{}x  + {}y + {} = 0'.format(self.a,self.b,self.c)
    
    def point_on_line(line,point):
        if line.a*point.x + line.b*point.y + line.c == 0:
            return 'lies on the line'
        else :
            return 'does not lies on the line'
    def shortest_distance(line,point):
        abs(line.a*point.x + line.b*point.y + line.c)/(line.a**2 + line.b**2)
        
```

```python
# Example of Method Overloading :
 # In Method overloading our user decide the method on the basis of parameters.
class shape:

    def area(self, radius):
        return 3.14*radius**2

    def area(self, l, b):
        return l*b

s = shape()
# s.area(45) # taking radius now because of one parameter
# but in python there is no way to use it .
# because python never allows Method overloading just like java don't allow multiple inheritance !
s.area(23, 45) # l * b because our parameters are two
```

```python
# but there is one hack to use the method overloading in python

class shape:

    def area(self, a, b = 0):
        if b == 0:   # because of condition we can kind of use method overloading !
            return 3.14*a**2
        else :
            return a + b

s = shape()
print(s.area(2))
print(s.area(3,4))
```

```python
# Examples for method overriding :

class Animal:
    def speak(self):
        print("The animal makes a sound")

class Dog(Animal):
    # Overriding the speak method
    def speak(self):
        print("The dog barks: Bhaauuu! Bhauuuu!")

class Cat(Animal):
    # Overriding the speak method
    def speak(self):
        print("The cat meows: Meow!")

# Usage
my_dog = Dog()
my_dog.speak()  # Output: The dog barks: bhaauuu! bhaauuuu!
```

```python
# more Example :
class Payment:
    def process_payment(self, amount):
        print(f"Processing a generic payment of ${amount}")

class CreditCardPayment(Payment):
    def process_payment(self, amount):
        # Specific logic for CC (connecting to bank API)
        print(f"Charging ${amount} to Credit Card. Applying 2% transaction fee.")

class CryptoPayment(Payment):
    def process_payment(self, amount):
        # Specific logic for Blockchain
        print(f"Transferring {amount} units via Blockchain wallet.")

# Both objects use the same method name, but behave differently
payments = [CreditCardPayment(), CryptoPayment()]
for p in payments:
    p.process_payment(100)
```

```python
from abc import ABC, abstractmethod
class BankApp(ABC):

    def database(self):
        print('connected to databases !')

    @abstractmethod
    def security(self):
        pass
```

```python
class mobileapp(BankApp):

    def mobile_login(self):
        print('login into mobile')

    def security(self):
        print('mobile Security ! ')
```

```python
mob = mobileapp()
```

```python
mob.security()
```

```python
mob.database()
```

```python
class MenuItem:
    def __init__(self, name, price):
        self.name = name
        self.price = price

class Restaurant:
    def __init__(self, name):
        self.name = name
        self.menu = []

    def add_item(self, name, price):
        new_item = MenuItem(name, price)
        self.menu.append(new_item)
        print(f" {name} ko menu mein add kar diya gaya.")

    def show_menu(self):
        if not self.menu:
            print("\n Menu abhi khali hai!")
        else:
            print(f"\n--- {self.name} Menu ---")
            for item in self.menu:
                print(f" {item.name}: {item.price}")

# --- Runtime Execution ---
my_restro = Restaurant("Apna Dhaba")

print(f"Welcome to {my_restro.name} Management System!")
print("Commands: 'add' (item add karne ke liye), 'show' (menu dekhne ke liye), 'exit' (bahar nikalne ke liye)")

while True:
    # Run time pe user se command lena
    command = input("\nCommand type karein: ").lower()

    if command == "add":
        n = input("Dish ka naam likhein: ")
        p = input("Price likhein: ")
        my_restro.add_item(n, p)

    elif command == "show":
        my_restro.show_menu()

    elif command == "exit":
        print("Program band ho raha hai. Bye!")
        break

    else:
        print(" Galat command! Sirf 'add', 'show' ya 'exit' likhein.")
```
